```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("processed.cleveland.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Replace '?' with NaN in my csv file
df.replace('?', np.nan, inplace=True)

# Display the dataset to check changes
df.head()
```

```python
# Now you can calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
correlation_matrix
```

```python
plt.xlabel("oldpeak")
plt.ylabel("num")
plt.scatter(df.oldpeak, df.num)
plt.show()
```

```python
plt.xlabel("chol")
plt.ylabel("thalach")
plt.plot(df.chol, df.thalach)
plt.grid()
plt.show()
```

```python
plt.xlabel("trestbps")
plt.ylabel("thalach")
plt.scatter(df.trestbps, df.thalach)
plt.grid()
plt.show()
```

```python
# Check for missing data
print(df.isnull().sum())
```

```python
# Convert 'ca' column to numeric, coercing errors to NaN
df['ca'] = pd.to_numeric(df['ca'], errors='coerce')

# Now fill in missing values for the 'ca' column with its mean
df['ca'] = df['ca'].fillna(df['ca'].mean())
```

```python
# Check for missing data
print(df.isnull().sum())
```

```python
# Convert 'thal' column to numeric, coercing errors to NaN
df['thal'] = pd.to_numeric(df['thal'], errors='coerce')

# Now fill in missing values for the 'ca' column with its mean
df['thal'] = df['thal'].fillna(df['thal'].mean())
```

```python
# Check for missing data
print(df.isnull().sum())
```
